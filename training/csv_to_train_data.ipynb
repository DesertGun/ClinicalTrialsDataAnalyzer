```python
def from_csv_to_train_data(datadf, label, colname):
    output = []
    for index, row in datadf.iterrows():
        char_start = 0
        if len(row[colname]) > 2:
            char_end = len(row[colname])-1
        else:
            char_end = len(row[colname])
        output_row = (row[colname], {"entities":[(char_start, char_end, label)]})
        output.append(output_row)
    return output
        
```

Cleaning the data

```python
data_conditions.dropna(inplace=True)
data_measurements.dropna(inplace=True)
```

```python
conditions_train = from_csv_to_train_data(data_conditions, "Condition", "name")
```
